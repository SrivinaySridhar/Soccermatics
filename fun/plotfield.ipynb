### Library to plot - [matplotlib](https://matplotlib.org/stable/tutorials/introductory/pyplot.html)

Importing `pyplot` as `plt`. Importing `Arc` from `patches`

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Arc
```

### Using the `plot` method to draw lines
`plt.plot()` takes 2 arguments of the following format: 

1. A list $[x_1,\ x_2]$ which specifies the start and end X coordinates
2. A list $[y_1,\ y_2]$ which specifies the start and end Y coordinates

This draws a line from $(x_1,\ y_1)$ as the starting point to $(x_2,\ y_2)$ as the ending point

```python
fig=plt.figure()
ax=fig.add_subplot(1,1,1)

plt.plot([0,0],[0,80], color="blue")
plt.plot([0,120],[80,80], color="orange")
plt.plot([120,120],[0,80], color="green")
plt.plot([0,120],[0,0], color="red")
plt.plot([60,60],[0,80], color="pink")

plt.show()
```

### Using the `circle` method to circles
`plt.circle()` takes 2 arguments of the following format: 

1. A tuple $(x,\ y)$ which specifies the X and Y coordinates of the centre of the circle
2. A float `n` which specifies the radius of the circle 

This draws a circle with centre $(x,\ y)$ and radius `n`

```python
#Create figure
fig=plt.figure()
ax=fig.add_subplot(1,1,1)

#Pitch Outline & Centre Line
plt.plot([0,0],[0,80], color="black")
plt.plot([0,120],[80,80], color="black")
plt.plot([120,120],[0,80], color="black")
plt.plot([0,120],[0,0], color="black")
plt.plot([60,60],[0,80], color="black")

#Assign circles to variables - do not fill the centre circle!
centreCircle = plt.Circle((60,40),9.15,color="red",fill=False)
centreSpot = plt.Circle((60,40),0.5,color="blue")

#Draw the circles to our plot
ax.add_patch(centreCircle)
ax.add_patch(centreSpot)

plt.show()
```

### Using the `Arc` method to draw arcs

`patches.Arc` takes 6 arguments of the following format:
1. A tuple $(x,\ y)$ which specifies the X and Y coordinates of the centre of the circle from which we are creating an arc
2. A float `h` which specifies the height of the circle (to disregard ellipse shapes)
3. A float `w` which specifies the width of the circle (to disregard ellipse shapes)
4. A float `a` which specifies the angle of rotation of the shape (anti-clockwise)
5. A float `theta1` which specifies the start location of the arc in degrees
6. A float `theta2` which specifies the end location of the arc in degrees

This draws an arc centered at $(x,\ y)$ with height `h`, width `w`, angle of rotation `a` degrees, starting from `theta1` degrees to `theta2` degrees

```python
#Demo Arcs

#Create figure
fig=plt.figure()
ax=fig.add_subplot(1,1,1)

#Pitch Outline & Centre Line
plt.plot([0,0],[0,80], color="black")
plt.plot([0,120],[80,80], color="black")
plt.plot([120,120],[0,80], color="black")
plt.plot([0,120],[0,0], color="black")
plt.plot([60,60],[0,80], color="black")

#Left Penalty Area
plt.plot([0,18],[62,62],color="black")
plt.plot([0,18],[18,18],color="black")
plt.plot([18,18],[18,62],color="black")

#Centre Circle/Spot
centreCircle = plt.Circle((60,40),9.15,color="black",fill=False)
centreSpot = plt.Circle((60,40),0.5,color="black")
ax.add_patch(centreCircle)
ax.add_patch(centreSpot)

#Create Arc and add it to our plot
leftArc = Arc((12,40),height=18.3,width=18.3,angle=0,theta1=310,theta2=50,color="black")

ax.add_patch(leftArc)

plt.show()
```

### Creating function `createPitch()` to create the pitch

```python
def createPitch():
    
    #Create figure
    fig=plt.figure()
    ax=fig.add_subplot(1,1,1)

    #Pitch Outline & Centre Line
    plt.plot([0,0],[0,80], color="black")
    plt.plot([0,120],[80,80], color="black")
    plt.plot([120,120],[0,80], color="black")
    plt.plot([0,120],[0,0], color="black")
    plt.plot([60,60],[0,80], color="black")
    
    #Left Penalty Area
    plt.plot([0,18],[62,62],color="black")
    plt.plot([0,18],[18,18],color="black")
    plt.plot([18,18],[18,62],color="black")
    
    #Right Penalty Area
    plt.plot([120,102],[62,62],color="black")
    plt.plot([120,102],[18,18],color="black")
    plt.plot([102,102],[18,62],color="black")
    
    #Left 6-yard Box
    plt.plot([0,6],[50,50],color="black")
    plt.plot([0,6],[30,30],color="black")
    plt.plot([6,6],[30,50],color="black")
    
    #Right 6-yard Box
    plt.plot([120,114],[50,50],color="black")
    plt.plot([120,114],[30,30],color="black")
    plt.plot([114,114],[30,50],color="black")
    
    #Prepare Circles
    centreCircle = plt.Circle((60,40),9.15,color="black",fill=False)
    centreSpot = plt.Circle((60,40),0.5,color="black")
    leftPenSpot = plt.Circle((12,40),0.5,color="black")
    rightPenSpot = plt.Circle((108,40),0.5,color="black")
    
    #Draw Circles
    ax.add_patch(centreCircle)
    ax.add_patch(centreSpot)
    ax.add_patch(leftPenSpot)
    ax.add_patch(rightPenSpot)
    
    #Prepare Arcs
    leftArc = Arc((12,40),height=18.3,width=18.3,angle=0,theta1=310,theta2=50,color="black")
    rightArc = Arc((108,40),height=18.3,width=18.3,angle=0,theta1=130,theta2=230,color="black")

    #Draw Arcs
    ax.add_patch(leftArc)
    ax.add_patch(rightArc)
    
    #Tidy Axes
    plt.axis('off')
    
    #Display Pitch
    plt.show()
```

```python
createPitch()
```
